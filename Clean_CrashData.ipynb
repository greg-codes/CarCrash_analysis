# Code

Import packages, define data directory path.

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns',None)
pd.set_option('display.width',None)
pd.set_option('display.max_colwidth',None)

crashpath = r'C:\CarCrash_analysis\CarCrashData'
```

Load .csv file, inspect columns.

```python
df = pd.read_csv(os.path.join(crashpath, 'CrashStatistics.csv'), low_memory=False)
```

Clean "LocalReportNumber" column.

```python
df.LocalReportNumber = df.LocalReportNumber.apply(lambda x: x.strip())
```

```python
usefulcols = ['LocalReportNumber', 'DocumentNumber', 'CrashSeverity', 
        'County', 'InCityVillageTownship',
       'CrashDateTime', 'Latitude',
       'Longitude', 'LocationRouteType', 'LocationRouteNumber',
       'LocationPrefix', 'LocationRoadName', 'LocationRoadType',
       'IntersectionOrApproachRelated',
       'NumberOfApproaches', 'WithinInterchangeArea',
       'LocationFirstHarmfulEvent', 'MannerOfCollision', 'Weather',
       'LightCondition', 'ActiveSchoolZoneRelated', 'WorkZoneRelated',
       'WorkersPresent', 'LawEnforcementPresentInWorkZone', 'WorkZoneType',
       'WorkZoneLocation', 
       'TotalTimeRoadwayClosed', 'RoadwayDivided',
       'DividedLaneTravelDirection', 'DividedMedianType', 'RoadContour',
       'RoadCondition', 'RoadSurface', 'TotalInjured', 'TotalKilled',
       'TotalMinutes', 'AnimalRelated', 'AnimalDeerRelated', 'AlcoholRelated',
       'DrugRelated', 'BicycleRelated', 'MotorCycleRelated', 'SpeedRelated',
       'PedestrianRelated', 'SemiTruckRelated', 'SmallTruckRelated',
       'YouthRelated', 'TeenRelated', 'DUI21Related', 'SeniorRelated',
       'FatalNotReportable', 'CommercialRelated', 'CommercialAtFault']
```

```python
df = df[usefulcols]
```

```python
df = df*1
```

Convert CrashDateTime column to datetime:

```python
df.CrashDateTime = pd.to_datetime(df.CrashDateTime, format="%m/%d/%Y %I:%M:%S %p")
```

```python
df.dtypes
```

Use one hot encoding on the categorical columns, then drop them.

```python
ohecols = ['CrashSeverity',
 'InCityVillageTownship',
 'LocationRouteType',
 'LocationRoadType',
 'LocationFirstHarmfulEvent',
 'MannerOfCollision',
 'Weather',
 'LightCondition',
 'WorkZoneType',
 'WorkZoneLocation',
 'DividedLaneTravelDirection',
 'DividedMedianType',
 'RoadContour',
 'RoadCondition',
 'RoadSurface']
```

```python
one_hot = pd.get_dummies(df[ohecols])
```

```python
df = df.drop(ohecols,axis=1)
df = df.join(one_hot)
```

Inspect final result.

```python
df.head()
```

Save cleaned df to disk.

```python
df.to_csv(os.path.join(crashpath, 'CrashStatistics_cleaned.csv'), index=False)
```
